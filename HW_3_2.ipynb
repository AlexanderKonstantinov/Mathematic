### 1. Задание (в программе)
Нарисуйте график функции:
y(x) = k∙cos(x – a) + b
для некоторых (2-3 различных) значений параметров k, a, b

```python
randfloat(1.1, 1.9)
```

```python
class RandomCosFunction:
        
    def __init__(self):
        
        self.k, self.a, self.b = randint(1, 9), randint(0, 9), randint(0, 9)
        
    def calc_func(self, x):
        return self.k * np.cos(x - self.a) + self.b
        
    def __str__(self):
        return "y(x)={}∙cos(x-{})+{}".format(self.k, self.a, self.b)
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 201)

f1 = RandomCosFunction() 
f2 = RandomCosFunction()
f3 = RandomCosFunction()

plt.plot(x, f1.calc_func(x), label=str(f1))
plt.plot(x, f2.calc_func(x), label=str(f2))
plt.plot(x, f3.calc_func(x), label=str(f3))
plt.gca().legend()

plt.xlabel('x')
plt.ylabel('y')
```

### 3. Задание (в программе)
1. Напишите код, который будет переводить полярные координаты в декартовы.

```python
def polar_to_decart(r, a):
    
    x = r * np.cos(a)
    y = r * np.sin(a)
   
    return x, y
```

```python
r_array, a_array = [], []

for _ in range(10):
    r_array.append(randint(1, 10))
    a_array.append(randfloat(-2 * np.pi, 2 * np.pi))
    
x, y = polar_to_decart(r_array, a_array)
```

3. Напишите код, который будет рисовать график отрезка прямой линии в полярных координатах.

```python
def plot_straight_segment():
    normal = 1
    alpha = np.pi / 2

    r = []
    phi = np.linspace(np.pi / 5, 3 * np.pi / 5)

    for phi_ in phi:
        r.append(normal / np.cos(phi_ - alpha))

    plt.polar(phi, r)    
```

```python
plot_straight_segment()
```

### 4. Задание
1. Решите систему уравнений:

    y = x^2 – 1
    
    exp(x) + x∙(1 – y) = 1

```python
x = np.linspace(-2, 3, 1000)
zero_index = np.argwhere(x == 0)
x = np.delete(x, zero_index)

def f1(x):
    return x**2 - 1

def f2(x):
    return (np.exp(x) - 1) / x + 1

def equations(p):
    x, y = p
    return y - x**2 + 1, np.exp(x) + x * (1 - y) - 1

p1 = fsolve(equations, (-2, 2))
p2 = fsolve(equations, (3, 8))

plt.plot(x, f1(x))
plt.plot(x, f2(x))
plt.scatter(
    [p1[0], p2[0]], 
    [p1[1], p2[1]],
    c='red'
)

plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
```

2. Решите систему уравнений и неравенств:
    
    y = x^2 – 1

    exp(x) + x∙(1 – y) > 1

```python
plt.plot(x, f2(x))

y1 = f1(x)

x3_1, y3_1 = [], []
x3_2, y3_2 = [], []

for _x in x:
    if _x >= p1[0]:
        break
    x3_1.append(_x)
    y3_1.append(f1(_x))
    
for _x in x[::-1]:
    if _x <= p2[0]:
        break
    x3_2.append(_x)
    y3_2.append(f1(_x))

plt.plot(x, f1(x))
plt.plot(x, f2(x))
plt.plot(x3_1, y3_1, c='r')
plt.plot(x3_2, y3_2, c='r', label='decision')

plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.gca().legend()
```
